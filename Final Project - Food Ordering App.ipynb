```python
class Resturant:
    def __init__(self,name):
        self.resturant_name    = name
        self.food              = {}
        self.food_id           = len(self.food)+1
        self.ordered_item      = []
        self.personal_details  = {}
        
        
    #Admin Functionalities
    
    
    def add_food_item(self):
        try:
            Name       = input("Enter food name :")
            Quantity   = float(input("Enter Food Quantity in value :"))
            Price      = float(input("Enter Price of food IN Rs :"))
            Discount   = float(input("Enter Discount here in Rs:"))
            Stock      = float(input("Enter the availabe stock in value :"))
            food_item  = {"Name":Name,"Quantity":Quantity,"Price":Price,"Discount":Discount,"Stock":Stock}
        
            self.food_id  = len(self.food)+1
            self.food[self.food_id] = food_item
            print("\n Food Item added Successfully !!")
            print("\n Newly Added item is",self.food,"\n")
            
        except Exception as e:
            print("\n Something Went wrong.Please try again...\n")
            
    def edit_food_item(self):
        try:
            f_id = int(input("Enter food id that you want to update :\n"))
            if f_id in self.food.keys():
                print("1.Update Food Name \n2.Update Quantity \n3.Update Price \n4.Update Discount \n5.Update Stock \n")
                y = input("Enter number to Update details :")
                if y == "1":
                    self.food[f_id]["Name"] = input("Enter new food name :")
                    print("\n New food name updated Successfully !! \n")
                elif y == "2":
                    self.food[f_id]["Quantity"] = float(input("Enter new qunatity in value :"))
                    print("\n New Qunatity updated Successfully !! \n")
                elif y == "3":
                    self.food[f_id]["Price"] = float(input("Enter new price in Rs :"))
                    print("\n New Price Updated Successfully !! \n")
                elif y == "4":
                    self.food[f_id]["Discount"] = float(input("Enter new Discount in Rs :"))
                    print("\n New Discount is updated Successfully !! \n")
                elif y == "5":
                    self.food[f_id]["Stock"] = float(input("Enter new Stock in value :"))
                    print("\n New Stock has been updated Successfully !! \n ")
                else:
                    print("\n Invalid Number \n")
            else:
                print("\n Incorrect Food Id ")
                
        except Exception as e:
            print("\n Something Went wrong.Please try again....\n")
            
    def view_food_item(self):
        print("List of food items : \n")
        if len(self.food) !=0:
            for i in self.food:
                print(f"Food ID : {i}")
                for j in self.food[i]:
                    print(j, ":", self.food[i][j])
                print()
                
        else:
            print("\n Sorry No Food items is Added !! \n")
            
    def delete_food_item(self):
        try:
            print("\n Warning !! Food item will delete Permanently !!")
            print("Enter Food ID")
            x = int(input())
            if x in self.food.keys():
                del self.food[x]
                print("\n Food item Deleted Successfully !!\n")
                print("Updated Food List :\n",self.food)
            else:
                print("\n Invalid Food ID \n")
        
        except Exception as e:
            print("\n Something went wrong please try again !!")
            
    #User Functionalities
        
    def user_registration(self):
        try:
            while True:
                Full_Name     = input("Enter your Full name :")
                Phone_Number  = int(input("Enter your 10 Digit mobile number :"))
                Email         = input("Enter your mail ID :")
                Address       = input("Enter your address :")
                Password      = input("Enter your password :")
                self.personal_details = {"Full Name":Full_Name,"Phone Number":Phone_Number,"Email":Email,"Address":Address,"Password":Password}
                print("\n Your account is created Successfully !! \n")
                print(f" Welcome to {self.resturant_name} Restaurant \n")
                print("User Details :")
                for i in self.personal_details:
                    print(i,":", self.personal_details[i])
                break
        except Exception as e:
            print("\n Something went wrong please try again !! \n")    
               
    def user_login(self):
        try:
            while True:
                print(f" Welcome to {self.resturant_name} Resturant \n")
                email    = input("Enter your Email ID :")
                password = input("Enter your Password :")
                if len(self.personal_details) != 0:
                    if email == self.personal_details["Email"] and password == self.personal_details["Password"]:
                        print("\n You are Login Successfully!!")
                        while True:
                            print("\n Enter the Below options \n")
                            print("\n1.Place New Order \n2.Check Order History \n3.Update Your Personal Detials \n4.Exit")
                            
                            detail = input()
                            if detail == "1":
                                self.placed_order()
                            elif detail == "2":
                                self.ordered_history()
                            elif detail == "3":
                                self.update_details()
                            elif details == "4":
                                break
                            else:
                                print("Invalid Number")
                              
                    else:
                        print("\n You have Entered Wrong Email or Password !! \n")
                    
                else:
                    print("\n There is no Account with this Email ID !! \n Please Create your Account first \n")
                    break
                break
        except Exception as e:
            print("\n Something went wrong Please try again!! \n")
            
    def placed_order(self):
        try:
            if len(self.food) != 0:
                menu = []
                for item in self.food:
                    menu.append([self.food[item]["Name"],self.food[item]["Quantity"],self.food[item]["Price"]])
                for i in menu:
                    print(i)
                while True:
                    print("\n Enter 1 to place order \n Enter 2 to exit \n")
                    x = input()
                    if x == "1":
                        print("\n Enter the food number that you want to Order in seperated comma \n")
                        y = input().split(",")
                        for i in y:
                            z = int(i)
                            if z <= len(menu):
                                self.ordered_item.append(menu[z-1])
                            else:
                                print("\n We Dont have this Food Item :",z)
                        print("List of selected food item :\n")
                        for j in self.ordered_item:
                            print(j)
                    elif x == "2":
                        break
                    else:
                        print("\n Invalid number")
            else:
                print("\n Sorry no food item are availabe now !! \n")
        except Except as e:
            print("\n Something Went wrong Please try again !! \n")
     
            
    def ordered_history(self):
        print("\n List of previously ordered Item :")
        for i in self.ordered_item:
            print(i)
    
    def update_details(self):
        try:
            while True:
                print("\n Select Below option to Update your Personal details \n")
                print("\n1.Name \n2.Phone Number \n3.Email \n4.Password \n5.Address \n6.Exit \n")
                x = input()
                if x == "1":
                    self.personal_details["Full Name"] = input("Enter your Updated Full name :")
                    print("\n Your Name Updated Successfully !! \n")
                elif x == "2":
                    self.personal_details["Phone Number"] = int(input("Enter your Updated 10 Digit Mobile Number :"))
                    print("\n Your Mobile Number Updated Successfully !! \n")
                elif x == "3":
                    self.personal_details["Email"] = input("Enter your Updated Email ID :")
                    print("\n Your Email ID Updated Successfully !! \n")
                elif x == "4":
                    self.personal_details["Password"] = input("Enter your Updated Password :")
                    print("\n Your Password Updated Successfully !! \n")
                elif x == "5":
                    self.personal_details["Address"] = input("Enter your Updated Address :")
                    print("\n Your Address Updated Successfully !! \n")
                
                elif x == "6":
                    break
                else:
                    print("\n You Entered Invalid Number \n")
                if x in ["1","2","3","4","5","6"]:
                    for i in self.personal_details:
                        print(i, ":" ,self.personal_details[i])
                else:
                    print("\n Please Enter Correct input \n")
        except Except as e:
            print("\n Something Went wrong Please try again...!! \n")
            
    
            #defining main function to run the code
    
try:
    def main():
        obj = Resturant("Ajay's")
        print(f"!! Welcome to {obj.resturant_name} Resturant !! \n")
        print("\n !!***************************************************!! \n")
        print("\n Hello Guys What do you want to Eat \n")
        print("\n !!***************************************************!! \n")
        while True:
            print("\n1. Admin \n2. User \n3. Exit \n")
            x = input()
            if x == "1":
                while True:
                    print("\n1. Add New Food Items \n2. Edit Food Items \n3. View Food Items \n4. Delet Food Items \n5. Exit")
                    y = input()
                    if y == "1":
                        obj.add_food_item()
                    elif y == "2":
                        obj.edit_food_item()
                    elif y == "3":
                        obj.view_food_item()
                    elif y == "4":
                        obj.delete_food_item()
                    elif y == "5":
                        break
                    else:
                        print("\n Invalid Number \n")
                    
            elif x == "2":
                while True:
                    print("\n Enter the Below Option \n")
                    print("\n1. Register \n2. Login \n3. Exit \n")
                    y = input()
                    if y == "1":
                        obj.user_registration()
                    elif y == "2":
                        obj.user_login()
                    elif y == "3":
                        break
                    else:
                        print("\n Invalid Number \n")
            elif x == "3":
                break
            else:
                print("\n Invalid Number \n")
except Exception as e:
    print("\n Something went wrong Please try again \n")
                
         
        #calling the main function
        
if __name__  == '__main__':
    main()
             
print("Thank You Sir To visit Our Resturant !!!")
                                
  
```
